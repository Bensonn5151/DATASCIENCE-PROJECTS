```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.neural_network import MLPClassifier
from sklearn.svm import SVC
from sklearn.metrics import mean_absolute_error, r2_score
from sklearn.metrics import accuracy_score, recall_score, precision_score, classification_report, confusion_matrix
```

```python
import time
lr=LogisticRegression()
dt=DecisionTreeClassifier()
kn=KNeighborsClassifier()
rf=RandomForestClassifier()
svc=SVC()
mlp=MLPClassifier(hidden_layer_sizes=(64,32), activation='relu',
                 max_iter=1000, random_state=42)
```

```python
regressors = {'Logistic Regression':lr, 'DecisionTree':dt, 'RandomForest':rf, 'MLP':mlp, 'SupportVector': svc}
```

```python
X=data.drop(['booking_complete','num_passengers', 'flight_hour', 'flight_day'],axis=1)
y=data['booking_complete']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.5, random_state=42)
```

```python
from sklearn.metrics import accuracy_score #used for categorical variables 
for name, reg in regressors.items():
    reg.fit(X_train,y_train)
    y_pred=reg.predict(X_test)
    accuracy = accuracy_score(y_test, y_pred)
    print(f"Regressor: {name}")
    print(f"accuracy: {accuracy}")
    print(classification_report(y_test, y_pred))
    print(confusion_matrix(y_test,y_pred))
```

The classification reports for both Decision Tree and Random Forest regressors show similar performance characteristics:

For Decision Tree:

Precision for class '0' is 0.86, meaning that when the model predicts '0', it is correct 86% of the time.
Recall for class '0' is 0.86, indicating that the model correctly identifies 86% of all instances of class '0' in the dataset.
F1-score for class '0' is 0.86, which is the harmonic mean of precision and recall.
For class '1', precision, recall, and F1-score are all around 0.22, indicating low performance in predicting class '1'.
The confusion matrix shows that the model correctly predicts a large number of instances for class '0' but struggles to correctly predict instances for class '1'.

Similarly, for Random Forest:

Precision, recall, and F1-score for class '0' are higher compared to class '1'.
The model performs relatively better in predicting class '0' instances compared to class '1' instances.
Overall, both models have a higher accuracy in predicting class '0' instances, but they struggle with class '1' due to its low recall, precision, and F1-score. This indicates a class imbalance issue, where the models are biased towards predicting the majority class. Applying techniques like oversampling, undersampling, or using algorithms that inherently handle class imbalance better could potentially improve the performance, especially for predicting class '1' instances.

```python
x=data.drop('booking_complete', axis=1)
y=data['booking_complete']
```

```python
for colname in x.select_dtypes('object'):
    x[colname], _ =x[colname].factorsize()
```

```python
x.dtypes
```

```python
from sklearn.feature_selection import mutual_info_classif
mi_scores=mutual_info_classif(x,y)
mi_scores=pd.Series(mi_scores,name='MI Scores', index=x.columns)
mi_scores=mi_scores.sort_values(ascending=False)
mi_scores
```

```python
def plot_scores(scores):
    scores=scores.sort_values(ascending=False)
    width=np.arange(len(scores))
    ticks=list(scores.index)
    plt.barh(width,scores)
    plt.yticks(width, ticks)
    plt.title("Mutual Information Scores")
plt.figure(dpi=100, figsize=(8,6))
plot_scores(mi_scores)
```
