```python
import pandas as pd
data=pd.read_csv('nigeria_houses_data.csv')
```

```python
data.head()
```

```python
#Using the groupby function to access the towns and count the occurence of each townssize_
size_town_price=data.groupby('town')['price'].size().sort_values()
size_town_price
```

```python
#Using the groupby function to access the mean of towns with their respective prices
mean_town_price=data.groupby('town')['price'].mean().sort_values()
mean_town_price
```

```python
#group the  data by the 'title' column and then count the occurrences of each 'title' & sorting them based on the count
size_title_price=data.groupby('title')['price'].size().sort_values()
size_title_price
```

```python
mean_title_price=data.groupby('title')['price'].mean().sort_values()
mean_title_price
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10,6))
size_title_price.plot(kind='bar', color='brown')
plt.xlabel("type")
plt.ylabel('price')
plt.title("Bar chart of houses based on type of building")
plt.show()
```

```python
plt.figure(figsize=(10,6))
mean_title_price.plot(kind='bar', color='brown')
plt.xlabel("type")
plt.ylabel('price')
plt.title("Bar chart of average house prices, based on type of building")
plt.show()
```

```python
import seaborn as sns
towns=data.groupby('town')['price'].mean().sort_values().reset_index()
plt.figure(figsize=(20,80))
sns.barplot(x='price',y='town', data=towns,palette='plasma', orient='h')
plt.title("Average house prices based on town and price")
plt.show()
```
